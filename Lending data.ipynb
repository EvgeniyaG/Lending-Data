```python
#import libraries

import pandas as pd
Location =  'loan.csv'
df = pd.read_csv(Location)
```

```python
df.head()
```

```python
df['loan_status'].unique()
```

```python
def default_to_number(x):
    if x=='Default':
        return 1
    else : 
        return 0
```

```python
df_short = df[['loan_status','loan_amnt','int_rate','annual_inc', 'term','purpose','emp_length']]
df_short.head()
```

```python
df_short['default_val'] = df_short['loan_status'].apply(default_to_number)
df_short.head()
```

```python
df_short.emp_length.unique()
```

```python
def emp_to_number(x):
    if x=='1 year':
        return 1
    if x=='5 years':
        return 5
    if x=='8 years':
        return 8
    if x=='10+ years':
        return 10
    if x=='n/a':
        return 0
    if x=='< 1 year':
        return 0.5
    if x=='7 years':
        return 7
    if x=='6 years':
        return 6
    if x=='4 years':
        return 4
    if x=='3 years':
        return 3
    if x=='9 years':
        return 9
    if x=='2 years':
        return 2
      
```

```python
df_short['emp_val'] = df_short['emp_length'].apply(emp_to_number)
df_short.head()
```

```python
df_short.term.unique()
```

```python
def term_to_number(x):
    if x==' 36 months':
        return 0
    if x==' 60 months':
        return 1
```

```python
df_short['term_val'] = df_short['term'].apply(term_to_number)
df_short.head()
```
